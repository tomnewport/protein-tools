# Load series data

```python
ss_ss2 = pandas.read_table('phyre2/query.ss2',engine='python', skiprows=2, header=None, sep=r"\s*")
ss_ss2.columns = ["pos","aa","ss2_max","ss2_coil","ss2_helix","ss2_strand"]
```

```python
ss_disopred = pandas.read_table('phyre2/query.ss_disopred',engine='python', skiprows=2, header=None, sep=r"\s*")
ss_disopred.columns = ["pos","aa","ss_disopred_max","ss_disopred_coil","ss_disopred_helix","ss_disopred_strand"]
```

```python
ss_ss = pandas.read_table('phyre2/query.ss',engine='python', skiprows=0, header=None, sep=r"\s*")
ss_ss.columns = ["pos","aa","ss_max","ss_coil","ss_helix","ss_strand"]
```

```python
ss_poing = pandas.read_table('phyre2/poing.ss',engine='python', skiprows=0, header=None, sep=r"\s*")
ss_poing.columns = ["pos","aa","poing_max","poing_coil","poing_helix","poing_strand"]
```

## Load transmembrane predictions

```python
tmhmm = pandas.read_table('tmhmm.txt',engine='python', skiprows=2, header=None, sep=r"\s*")
tmhmm.columns = ["pos", "aa","tmhmm_inside","tmhmm_membr", "tmhmm_outside"]
```

## Load coiled coil predictions

```python
coilsmdk = pandas.read_table('coils_mdk.txt',engine='python', skiprows=5, header=None, sep=r"\s*")
coilsmdk.columns = ["pos", "aa", "coilsmdk_pos", "coilsmdk1", "temp1", "coilsmdk2", "temp2", "coilsmdk3"]
del coilsmdk['temp1']
del coilsmdk['temp2']

coilsmtidk = pandas.read_table('coils_mtidk.txt',engine='python', skiprows=5, header=None, sep=r"\s*")
coilsmtidk.columns = ["pos", "aa", "coilsmtidk_pos", "coilsmtidk1", "temp1", "coilsmtidk2", "temp2", "coilsmtidk3"]
del coilsmtidk['temp1']
del coilsmtidk['temp2']
```

# Load annotation data

```python
def clean_dict(in_dict, discard):
    out_dict = {}
    for key, value in in_dict.items():
        if key not in discard and value is not None:
            if isinstance(value, (np.float64, int, float, complex)):
                if not np.isnan(value):
                    out_dict[key] = value
            else:
                if value not in ["-", "_"]:
                    out_dict[key] = value
    return out_dict

def summarise(row,as_json=True):
    discard = ['ip_length', 
                'ip_status', 
                'ip_letter',
                'ip_id', 
                'ip_date', 
                'start', 
                'end', 
                'phyre2_scop_PDBTitle_x', 
                'phyre2_scop_PDBTitle_y',
                'phyre2_scop_PDB header_y',
                'phyre2_scop_PDB header_x',
                'phyre2_scop_Chain_x',
                'phyre2_scop_Chain_y',
                'phyre2_scop_PDB Molecule_x',
                'phyre2_scop_PDB Molecule_y',
                'phyre2_scop_Chain']
    if as_json:
        return json.dumps(clean_dict(row.to_dict(), discard))
    else:
        return clean_dict(row.to_dict(), discard)
def titlise(row):
    maxlen = 40
    titles = ["phyre2_scop_Family", "phyre2_scop_PDB Molecule", "ip_db_feature", "ip_db_desc"]
    title = []
    for key, value in summarise( row, as_json=False ).items():
        if key in titles:
            v = value.strip(";-_ ");
            if len(v) > maxlen:
                v = v[:maxlen - 3] + "..."
            v = v[:1].upper() + v[1:]
            title.append(v)
    title.sort(key=len)
    return " - ".join(title)
def hashify(row):
    return hashlib.sha256(json.dumps(row.to_dict()).encode('utf-8')).hexdigest()[:16]
```

```python
domains_phyre = pandas.read_table('phyre2/crudelist',engine='python', skiprows=0, header=None, sep=r"\s*")
domains_phyre.columns=['phyre2_order', 'phyre2_template','phyre2_match', 'phyre2_coverage','start','end']
domains_phyre['phyre2_match_type'] = domains_phyre['phyre2_template'].str[0]
domains_phyre['phyre2_match_pdb_id'] = domains_phyre['phyre2_template'].str[1:5]
domains_phyre['phyre2_match_chain'] = domains_phyre['phyre2_template'].str[5]
domains_phyre['phyre2_match_domain'] = domains_phyre['phyre2_template'].str[6]
domains_phyre['phyre2_match_info_location'] = domains_phyre['phyre2_template'].str[0:7]
```

```python
def dict_from_scop(scop_file, prefix):
    dict_return = {}
    sf = open(scop_file,'r')
    alltext = ""
    for line in sf:
        alltext += line
    #remove line breaks:
    alltext = alltext.replace('\n','').replace('<br>','')
    soup = BeautifulSoup(alltext)
    parts = soup.find('body').contents
    for i in range(0,len(parts)//2):
        key = parts[i*2].contents[0].replace(':','').strip()
        value = parts[(i*2)+1].replace(':','').strip()
        dict_return[prefix+key] = value
    return dict_return

scops = {}
for scop in domains_phyre['phyre2_match_info_location']:
    scops[scop] = dict_from_scop('phyre2/'+scop+'.scop', "phyre2_scop_")
domain_info = pandas.DataFrame.from_dict(scops).transpose()
domains_phyre = domains_phyre.merge(domain_info, right_index=True, left_on='phyre2_match_info_location')
domains_phyre["summary"] = domains_phyre.apply(summarise, axis=1)
domains_phyre["title"] = domains_phyre.apply(titlise, axis=1)
domains_phyre["source"] = "Phyre2"
domains_phyre["hash"] = domains_phyre.apply(hashify, axis=1)
```

```python
domains_interpro = pandas.read_table('interpro.txt', sep='\t', names=['ip_status', 'ip_id', 'ip_length', 'ip_database', 'ip_db_feature', 'ip_db_desc', 'start', 'end', 'ip_e-value', 'ip_letter', 'ip_date', 'ip_interpro_id', 'ip_interpro_name', 'ip_GO', 'ip_unknown'], engine = "python")
```

```python
domains_interpro["summary"] = domains_interpro.apply(summarise, axis=1)
domains_interpro["title"] = domains_interpro.apply(titlise, axis=1)
domains_interpro["source"] = "InterPro"
```

# Combine series data

```python
series = ss_poing.merge(ss_ss).merge(ss_ss2).merge(ss_disopred).merge(diso_diso).merge(diso_metaprdos).merge(tmhmm).merge(coilsmdk).merge(coilsmtidk)
```

```python
series.to_csv("../series.csv",index=False)
```

```python
annotations = domains_interpro[["hash", "title","start","end","summary","source"]].append(domains_phyre[["hash", "title","start","end","summary","source"]])
```
